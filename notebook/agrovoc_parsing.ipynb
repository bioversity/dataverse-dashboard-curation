```python
import json
from urlparse import urlparse
```

```python
#### AGOVOC mapping
agrovoc_data = []

with open('agrovocLabels.json') as data_file:
    agrovoc = json.load(data_file)
    results = agrovoc["results"]["bindings"]
    for entry in results:
        # Parse the URI, split by slashes and consider the last part as the AGRVOC term ID
        term = entry["label"]["value"]
        agrovoc_term_ID = urlparse(entry["uri"]["value"]).path.split("/")[3]
        
        agrovoc_data.append({"ID": agrovoc_term_ID , "term": term})
        #print(term)

print "AGROVOC DATASET:\n\n", json.dumps(agrovoc_data)
```
