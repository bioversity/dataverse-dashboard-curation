```python
import json
from urlparse import urlparse
```

```python
def keys_exists(element, *keys):
    '''
    Check if *keys (nested) exists in `element` (dict).
    '''
    if type(element) is not dict:
        raise AttributeError('keys_exists() expects dict as first argument.')
    if len(keys) == 0:
        raise AttributeError('keys_exists() expects at least two arguments, one given.')

    _element = element
    for key in keys:
        try:
            _element = _element[key]
        except KeyError:
            return False
    return True
```

```python
#### AGOVOC mapping
agrovoc_data = []

with open('agrovocLabels.json') as data_file:
    agrovoc = json.load(data_file)
    results = agrovoc["results"]["bindings"]
    for entry in results:
        # Parse the URI, split by slashes and consider the last part as the AGRVOC term ID
        term = entry["label"]["value"]
        agrovoc_term_ID = urlparse(entry["uri"]["value"]).path.split("/")[3]
        
        agrovoc_data.append({"ID": agrovoc_term_ID , "term": term})
        #print(term)

print "AGROVOC DATASET:\n\n", json.dumps(agrovoc_data)
```

```python
#### Bioversity dataset mapping
bioversity_data = []

with open('BioversityDataset_sample.json') as data_file:
    bio_data = json.load(data_file)
    
    for g in bio_data:
        for fields in g["datasetVersion"]["metadataBlocks"]["citation"]["fields"]:
            for h in fields["value"]:
                
                # Check if is object
                #if isinstance(h, dict):
                    # Check if the object "h" contains given key
                    #if(format(keys_exists(h, "keywordVocabulary"))):
                        
                if "keywordVocabulary" in h:
                    #print json.dumps(h["keywordValue"]["value"])
                    bioversity_data.append(h["keywordValue"]["value"])

print "Keywords in Bioversity dataset:", json.dumps(bioversity_data), "\n"

for terms in bioversity_data:
    for i, obj in enumerate(agrovoc_data):
        if terms in obj['term']:
            print "Found '" + obj["term"] + "' in AGROVOC dataset", "with ID:", obj["ID"]
        
```
